```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('HT_2013-2020.csv')
df
```

```python
# check data for null values
df.isnull().sum()
```

```python
# change name of columns to lowercase
df.columns = [col.lower() for col in df.columns]
# 
df
```

```python
# check data for null values
df.isnull().sum()
```

```python
# drop columns 'pub_agency_unit', 'county_name'
df.drop(['pub_agency_unit', 'county_name'], axis=1, inplace=True)
```

```python
# aggrigate data by location, year, and offense_subcat_name
df_agg = df.groupby(['region_name', 'data_year', 'offense_subcat_name'])
```

```python
# unstack the data
df_agg = df_agg.size().unstack(fill_value=0)
df_agg
```

```python
# add new column named 'next_year_total'
df_agg['next_year_total'] = df_agg.sum(axis=1)
df_agg
```

```python
# select only the midwest region from df
df_midwest = df_agg.loc['Midwest']
df_northern = df_agg.loc['Northeast']
df_southern = df_agg.loc['South']
df_western = df_agg.loc['West']
```

```python
# create four subplots
fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(10, 10))
# subplot 1
ax[0].set_title('Midwest')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of Cases')
ax[0].set_xticks(np.arange(2013, 2021, 1))
ax[0].set_xticklabels(np.arange(2013, 2021, 1))
ax[0].set_yticks(np.arange(0, df_midwest.max().max() + 100, 100))
ax[0].set_yticklabels(np.arange(0, df_midwest.max().max() + 100, 100))
ax[0].plot(df_midwest.index, df_midwest.iloc[:, 0], color='blue', label='Midwest')
ax[0].plot(df_midwest.index, df_midwest.iloc[:, 1], color='red', label='Midwest', linestyle='--')
# subplot 2
ax[1].set_title('Northeast')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number of Cases')
ax[1].set_xticks(np.arange(2013, 2021, 1))
ax[1].set_xticklabels(np.arange(2013, 2021, 1))
ax[1].set_yticks(np.arange(0, df_northern.max().max() + 100, 100))
ax[1].set_yticklabels(np.arange(0, df_northern.max().max() + 100, 100))
ax[1].plot(df_northern.index, df_northern.iloc[:, 0], color='blue', label='Northeast')
ax[1].plot(df_northern.index, df_northern.iloc[:, 1], color='red', label='Northeast', linestyle='--')
# subplot 3
ax[2].set_title('South')
ax[2].set_xlabel('Year')
ax[2].set_ylabel('Number of Cases')
ax[2].set_xticks(np.arange(2013, 2021, 1))
ax[2].set_xticklabels(np.arange(2013, 2021, 1))
ax[2].set_yticks(np.arange(0, df_southern.max().max() + 100, 100))
ax[2].set_yticklabels(np.arange(0, df_southern.max().max() + 100, 100))
ax[2].plot(df_southern.index, df_southern.iloc[:, 0], color='blue', label='South')
ax[2].plot(df_southern.index, df_southern.iloc[:, 1], color='red', label='South', linestyle='--')
plt.legend()
# subplot 4
ax[3].set_title('West')
ax[3].set_xlabel('Year')
ax[3].set_ylabel('Number of Cases')
ax[3].set_xticks(np.arange(2013, 2021, 1))
ax[3].set_xticklabels(np.arange(2013, 2021, 1))
ax[3].set_yticks(np.arange(0, df_western.max().max() + 100, 100))
ax[3].set_yticklabels(np.arange(0, df_western.max().max() + 100, 100))
ax[3].plot(df_western.index, df_western.iloc[:, 0], color='blue', label='West')
ax[3].plot(df_western.index, df_western.iloc[:, 1], color='red', label='West', linestyle='--')
# increase the space between subplots
plt.subplots_adjust(hspace=0.5)
```
